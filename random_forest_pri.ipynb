# Random Forest

```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
features = pd.read_csv('modified_imputed_data.csv')
```

```python
#features = features.dropna()
#features.isnull().values.any()
features.shape
```

```python
round(features.describe(), 2)
```

```python
features.head()
```

```python
f = features.drop('Unnamed: 0', axis = 1)
f = f.drop('Unnamed: 0.1', axis = 1)
f.head()
```

```python
test = f[f['TimePeriod'] >= 2017]
train = f[f['TimePeriod'] < 2017]
test.to_csv("test.csv")
print(test.shape)
print(train.shape)
test.head()
```

```python
test_labels = test['target_variable']
test_features = test.drop('Footnote', axis = 1)
#test_features = test_features.drop('Indicator', axis = 1)
test_features = test_features.drop('target_variable', axis = 1)
```

```python
train_labels = train['target_variable']
train_features = train.drop('Footnote', axis = 1)
#train_features = train_features.drop('Indicator', axis = 1)
train_features = train_features.drop('target_variable', axis = 1)
```

```python
print(test_features.shape)
```

```python
round(test_labels.describe(), 2)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Instantiate random forest and train on new features
from sklearn.ensemble import RandomForestClassifier

### TODO ###
rf_exp = RandomForestClassifier(n_estimators= 200, random_state=19, n_jobs=-1)
rf_exp.fit(train_features, train_labels)
```

```python
# Make predictions on test data
from sklearn.metrics import accuracy_score

predictions = rf_exp.predict(test_features)
print(predictions)
accuracy_score(test_labels, predictions)
```

```python
# creating a confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(test_labels, predictions)
```

```python
# Get numerical feature importances
### TODO ###
# List of features for later use
feature_list = list(train_features.columns)
#print(feature_list)

importances = list(rf_exp.feature_importances_)
#print(importances)

# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]

# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
